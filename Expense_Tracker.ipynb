```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/synthetic_data_expenses.csv')
data.head()
```

```python
data.tail()
```

```python
data.info()
```

## **Expenses Distribution**

```python
## Expenses Distribution
plt.figure(figsize=(12,8))
plt.pie(data['amount'].groupby(data['category']).sum(),
        labels=data['category'].unique(), autopct="%1.1f%%")
plt.title("Expense Distribution By Category")
plt.show()
```

```python
sns.countplot(x='category', data=data, order=data['category'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Count Of Expenses Per Category')
plt.show()
```

##**Spending Trends Over TIme**

```python
data['date'] = pd.to_datetime(data['date'])
## Day of week
data['day_of_week'] = data['date'].dt.day_name()

## Monthly
data['month'] = data['date'].dt.month

## Week of month
data['week_of_month'] = data['date'].apply(lambda d: (d.day-1) // 7 + 1)
```

```python
data['date'] = pd.to_datetime(data['date']) ## To ensure correct date format

data.set_index('date',inplace=True) # Set date as index

data['amount'].resample('M').sum().plot(figsize=(10,6))
plt.title("Monthly Expenses")
plt.xlabel('Date')
plt.ylabel("Total Amount")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='date',y='amount',data=data,hue='category')
plt.xticks(rotation=45)
plt.title("Expenses Over Time")
plt.show()
```
